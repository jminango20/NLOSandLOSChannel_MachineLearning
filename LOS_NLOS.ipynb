# System Model

The channel considered is a Ultra-Wide Band (UWB) channel model due to the potential strength of this channel lies in its use of extremely wide transmission bandwidths, which results in desirable capabilities including accurate position location and ranging, lack of significant fading, high multiple access capability, covert communications, and possible easier material penetration [1] [2].

The large bandwidth of UWB channel can give rise to new effects. Thus, only few multipath components overlap within each resolvable delay bin, so that the central limit theorem is no longer applicable, and the amplitude fading statistics are
no longer Rayleigh. Furthermore, UWB waveforms can be up to 7.5 GHz wide, for example, paths separated by more than
about 133 pico seconds (ps) (equivalent to 4cm pathlength difference) can be individually resolved at the receiver. 


### Channel Model

#### Power Delay Profile

Here, we have considered that, in time domain, the impulse response of UWB channel is given by:

$
h(t) = f(t)*\sum_{l=1}^{L}\sum_{k=1}^{K}\alpha_{k,l}\delta(t-T_{l}-\tau_{k,l}), (1)
$

which represents path frequency dependence, where $L$ and $K$ denote the total number of clusters and rays respectively, $\alpha_{k,l}$ is the cofficiency of the $k$-th ray in the $l$-th cluster, and $\tau_{k,l}$ is the arrival time of $k$-th ray relative to $l$-th cluster arrival time $T_l$. 

In frequency domain, (1) can be rewritten as: 

$
H(\omega) = F(\omega)\sum_{l=1}^{L}\sum_{k=1}^{K}\alpha_{k,l}\hspace{0.1cm}\mathrm{exp}(-j\omega(T_{l}+\tau_{k,l})), (2)
$

where $F(\omega)$ denotes the frecuency dependence of rays arrivals which is given by:

$
F(\omega) = C_0(\omega/\omega_0)^{-\kappa}, (3)
$

where $C_0$ is a constant, $\kappa$ is the frequency dependence of the pathloss, and $\omega_0$ is the reference frequency. 

Moreover, using Taylor series, $F(\omega)$ can be expanded as: 
$
F(\omega) \approx F(\omega_c) + F'(\omega_c)(\omega-\omega_c) 
$, where $\omega_c$ denotes the center frecuency.

According to [3], the distribution of the cluster arrival times is given by a Poisson processes: 

$
p(T_l|T_{l-1}) = \Lambda_l \mathrm{exp}[-\Lambda_l(T_l - T_{l-1})], \hspace{0.2cm} l>0
$

where $\Lambda_l$ is the cluster rate.

The distribution of $\tau_{k,l}$ is modelled with a mixture of two Poisson processes as follows:

$
p(\tau_{k,l}|\tau_{(k-1),l}) = \beta\lambda_1 \mathrm{exp}[-\lambda_1( \tau_{k,l}-\tau_{(k-1),l} )] + (1-\beta)\lambda_2 \mathrm{exp}[-\lambda_2( \tau_{k,l}-\tau_{(k-1),l} )], \hspace{0.2cm} k>0
$

where $\beta$ is the mixture probability and $\lambda_1$ and $\lambda_2$ are the ray arrival rates.

Following, we have the determination of the cluster powers and cluster shapes. The power delay profile (mean power of the
different paths) is exponential within each cluster:
$
E[\left |\alpha_{k,l} \right |^2]=\frac{\Omega_l}{\gamma_l}\mathrm{exp}(-\tau_{k,l}/\gamma_l)
$

where $\Omega_l$ is the integrated energy of the $l$-th cluster, and $\gamma_l$ is the intra-cluster decay time constant.

The cluster decay rates are found to depend linearly on the arrival time of the cluster as:
$
\gamma_l \propto k_{\gamma}T_{l} + \gamma_0,
$

where $k_{\gamma}$ is the increase of the decay constant with delay.

The mean (over the cluster shadowing) mean (over the small-scale fading) energy (normalized to $\gamma_l$), of the $l$-th cluster follows in general an exponential decay:

$
10\log(\Omega_l)=10\log(\mathrm{exp}(-T_l/\Gamma)) + M_{cluster},
$

where $M_{cluster}$ is a normally distributed variable with standard deviation $\sigma_{cluster}$ around it.


#### Small-scale fading

The fading amplitude $\alpha_{k,l}$ follows a Nakagami-m distribution with parameters $(\Omega,m)$ according to:

\begin{equation}
p(\alpha_{k,l})=\frac{2}{\Gamma(m_{k,l})}(\frac{m_{k,l}}{\Omega_{k,l}})^{m_{k,l}}\alpha_{k,l}^{2m_{k,l}-1}\mathrm{exp}(-\frac{m_{k,l}}{\Omega_{k,l}}\alpha_{k,l}^{2}),
\end{equation}

where $\Gamma(\cdot)$ is the Gamma function, $m\geq 1/2$ is the Nakagami $m$-factor which is modeled as a lognormally distribution random variable and $\Omega_{k,l}$ is the mean-square value of the amplitude.

The parameter $\Omega_{k,l}$ corresponds to the mean power and its delay dependence is thus given by the power delay profile above. The $m$âˆ’factor has a mean $\mu_m$ and standard deviation $\sigma_m$. Both of these can have a delay dependence:
\begin{equation}
\mu_m(\tau) = m_0 - k_m\tau
\end{equation}
\begin{equation}
\sigma_m(\tau) = \hat{m}_0 - \hat{k}_m\tau
\end{equation}

For the first component of each cluster, the Nakagami $m$ factor is modeled differently. It is assumed to be deterministic and independent of delay:
\begin{equation}
m = \tilde{m}_0.
\end{equation}

#### Measurements

- __Mean Excess Delay__ (MED) is given by:
\begin{equation}
\tau_{MED} =\frac{\sum_{l=1}^{L}\sum_{k=1}^{K}(T_{l}+\tau_{k,l})\left | \alpha_{k,l} \right |^2}{\sum_{l=1}^{L}\sum_{k=1}^{K} \left | \alpha_{k,l} \right |^2}.
\end{equation}

- __Root Mean Square__ (RMS) delay spread is computed by:
\begin{equation}
\tau_{RMS} =\frac{\sum_{l=1}^{L}\sum_{k=1}^{K}((T_{l}+\tau_{k,l})-\tau_{MED})^2\left | \alpha_{k,l} \right |^2}{\sum_{l=1}^{L}\sum_{k=1}^{K} \left | \alpha_{k,l} \right |^2}.
\end{equation}

The simulation results have considered the realization of $10^3$ UWB channels.

```python
%%html 
<img src = "Delay_RMS.png", width=400, height=400>
```

- __K Factor__ in dB is computed by:
\begin{equation}
K_{factor} =10\log\left (\frac{\left | \alpha_{1,1} \right |^2}{\sum_{l=2}^{L}\sum_{k=2}^{K} \left | \alpha_{k,l} \right |^2} \right ).
\end{equation}

#### We prepare the data for the machine learning classification.

Note that we have taken  the 80% of the data for training and 20% for validation. That is, from the 5000 UWB channel measurements, 4000 are for training the algorithm and 1000 for validate the algorithm.

```python
#Dividir os elementos
X = df.drop( ['Tipo','index'] , axis=1 )
#X.head()
y = df['Tipo']

## SEPARO OS ELEMENTOS PARA TREINAMENTO E TESTE
#80% training set and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)

print('size our training "X" is', X_train.shape)
print('\n')
print('size our testing "X" is', X_test.shape)
print('\n')
print('size our training "y" is', y_train.shape)
print('\n')
print('size our testing "y" is', y_test.shape)
```

### We compare the different algorithms for classification for LOS and NLOS channels

```python
MLA = [
    #Ensemble Methods
    ensemble.AdaBoostClassifier(),
    ensemble.BaggingClassifier(),
    ensemble.ExtraTreesClassifier(),
    ensemble.GradientBoostingClassifier(),
    ensemble.RandomForestClassifier(),

    #Gaussian Processes
    gaussian_process.GaussianProcessClassifier(),
    
    #GLM
    linear_model.LogisticRegressionCV(),
    linear_model.PassiveAggressiveClassifier(),
    linear_model. RidgeClassifierCV(),
    linear_model.SGDClassifier(),
    linear_model.Perceptron(),
    
    #Navies Bayes
    naive_bayes.BernoulliNB(),
    naive_bayes.GaussianNB(),
    
    #Nearest Neighbor
    neighbors.KNeighborsClassifier(),
    
    #SVM
    svm.SVC(probability=True),
    svm.NuSVC(probability=True),
    svm.LinearSVC(),
    
    #Trees    
    tree.DecisionTreeClassifier(),
   #tree.ExtraTreeClassifier(),
    
    ]
```

```python
#Normalizo usando o StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

##### We train the data into the model and calculate the performance.

Note that the performance is obtained by the validation data.

```python
MLA_std_columns = []
MLA_std_compare = pd.DataFrame(columns = MLA_std_columns)

row_index = 0

for alg in MLA:   
    predicted_std = alg.fit(X_train_std, y_train).predict(X_test_std)
    fp_std, tp_std, th_std = roc_curve(y_test, predicted_std)
    MLA_std_name = alg.__class__.__name__
    MLA_std_compare.loc[row_index,'MLA Name'] = MLA_std_name
    MLA_std_compare.loc[row_index, 'MLA Train Accuracy'] = round(alg.score(X_train_std, y_train), 4)
    MLA_std_compare.loc[row_index, 'MLA Test Accuracy'] = round(alg.score(X_test_std, y_test), 4)
    MLA_std_compare.loc[row_index, 'MLA Precission'] = precision_score(y_test, predicted_std)
    MLA_std_compare.loc[row_index, 'MLA AUC'] = auc(fp_std, tp_std)
    row_index += 1
    
MLA_std_compare.sort_values(by = ['MLA Test Accuracy'], ascending = False, inplace = True)    
MLA_std_compare
```

##### We plot the test or validation performance of the different algorithms, which are ordered from the best to the worst.

```python
plt.subplots(figsize=(15,6))
sns.barplot(x="MLA Name", y="MLA Test Accuracy",data=MLA_std_compare,palette='hot',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('MLA Validation Accuracy Comparison')
plt.show()
```

##### Now, we use another performance criterion which is:

AUC (Area Under The Curve) ROC (Receiver Operating Characteristics) curve. It is one of the most important evaluation metrics for checking any classification modelâ€™s performance.

- AUC - ROC curve is a performance measurement for classification problem at various thresholds settings. 
  -- ROC is a probability curve and 
  -- AUC represents degree or measure of separability. 
  
AUC tells how much model is capable of distinguishing between NLOS and LOS classes. Higher the AUC, better the model is at predicting or distinguishing between NLOS as NLOS and LOS as LOS. 


```python
index = 1

for alg in MLA:  
    predicted_std = alg.fit(X_train_std, y_train).predict(X_test_std)
    fp, tp, th = roc_curve(y_test, predicted_std)
    roc_auc_mla = auc(fp, tp)
    MLA_name = alg.__class__.__name__
    plt.plot(fp, tp, lw=2, alpha=0.3, label='ROC %s (AUC = %0.2f)'  % (MLA_name, roc_auc_mla))
    index+=1

plt.title('ROC Curve comparison')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')    
plt.show()
```

###### NOTE: 
An excellent model has AUC near to the 1 which means it has good measure of separability. A poor model has AUC near to the 0 which means it has worst measure of separability. And when AUC is 0.5, it means model has no class separation capacity whatsoever.

Based on validation accuracy comparison, AUC and ROC, we determined that the __Gradient Boosting Classifier__ achieves a better performance.

### We work over Gradient Boosting Classifier

- Firstly, we introduce the concept of confusion matrix.
  -- It is a performance measurement for machine learning classification problem where output can be two or more classes.
  -- It is a table with 4 different combinations of predicted and actual values.

```python
%%html 
<img src = "conf_matrix.png", width=400, height=200>
```

     --- True Positive: You predicted positive and it is true.
     --- True Negative: You predicted negative and it is true.
     --- False Positive (Type 1 Error): You predicted positive and it is false.
     --- False Negative (Type 2 Error): You predicted negative and it is false.

Note that predicted values are described as Positive and Negative and actual values as True and False.

In order to improve the performance of a given algorithm, we need to reduce False Positive and/or False Negative.

In the following, we compute the confusion matrix for the selected __Gradient Boosting Classifier__


```python
# import machine learning algorithms
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

# train with Gradient Boosting algorithm
gb = GradientBoostingClassifier()
gb.fit(X_train_std, y_train)

y_predict_gb = gb.predict(X_test_std)

#Matriz de confusion
cm_gb = np.array(confusion_matrix(y_test,y_predict_gb,labels=[0,1]))
confusion_gb = pd.DataFrame(cm_gb,index=['LOS','NLOS'],
                         columns=['predicted_LOS','predicted_NLOS'])

sns.heatmap(confusion_gb,annot=True,fmt="d")
print(classification_report(y_test, y_predict_gb))

#Medicion Accuracy
acu_score_sgd = accuracy_score(y_test, y_predict_gb, normalize=True, sample_weight=None)
acu_score_sgd.mean()
```

- We test different learning rates, which is an important parameter, for the __Gradient Boosting Classifier__ algorithm and we print the accuracy score.

```python
# train with Gradient Boosting algorithm
# compute the accuracy scores on train and validation sets when training with different learning rates

learning_rates = [0.05, 0.1, 0.15, 0.20, 0.25, 0.30]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(criterion='friedman_mse', init=None,
              learning_rate=learning_rate, loss='deviance', max_depth=3,
              max_features=None, max_leaf_nodes=None,
              min_impurity_decrease=0.0, min_impurity_split=None,
              min_samples_leaf=1, min_samples_split=2,
              min_weight_fraction_leaf=0.0, n_estimators=100,
              n_iter_no_change=None, presort='auto', random_state=None,
              subsample=1.0, tol=0.0001, validation_fraction=0.1,
              verbose=0, warm_start=False)
    gb.fit(X_train_std, y_train)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (validation): {0:.3f}".format(gb.score(X_test_std, y_test)))
    print()
```

- From the above results. We can see that a _learning rate_ of _0.2_ produces better _accuracy score_ of _0.814_. Thus, we select this learning rate.


- Following, we compute the confusion matrix with this new learning rate.

```python
gb = GradientBoostingClassifier()

gb = GradientBoostingClassifier(criterion='friedman_mse', init=None,
              learning_rate=0.2, loss='deviance', max_depth=3,
              max_features=None, max_leaf_nodes=None,
              min_impurity_decrease=0.0, min_impurity_split=None,
              min_samples_leaf=1, min_samples_split=2,
              min_weight_fraction_leaf=0.0, n_estimators=100,
              n_iter_no_change=None, presort='auto', random_state=None,
              subsample=1.0, tol=0.0001, validation_fraction=0.1,
              verbose=0, warm_start=False)

gb.fit(X_train_std, y_train)
y_predict_gb = gb.predict(X_test_std)

#Matriz de confusion
cm_gb = np.array(confusion_matrix(y_test,y_predict_gb,labels=[0,1]))
confusion_gb = pd.DataFrame(cm_gb,index=['LOS','NLOS'],
                         columns=['predicted_LOS','predicted_NLOS'])

sns.heatmap(confusion_gb,annot=True,fmt="d")
print(classification_report(y_test, y_predict_gb))

#Medicion Accuracy
acu_score_sgd = accuracy_score(y_test, y_predict_gb, normalize=True, sample_weight=None)
acu_score_sgd.mean()
```

- From this confusion matrix, we note that the number of False Positives (FP) is reduced from 104 to 95, which produces an improvement in the performance of __Gradient Boosting Classifier__ algorithm.
